I am trying to break down individual performance for a player within the game. To do this, I am going to try and use a fantasy lcs score system to measure this. The link to the calculations can be found here https://medium.com/@jonny_32483/fantasy-lcs-scoring-explained-3fa9cae07e08

impact score is $(kills*3)-deaths +(assists*1.5)+(cs*.02)$
this is extremely basic, and really just a proof of concept for a tougher model later.
just for testing purposes, will upgrade later. 

```python
testingdf = pd.DataFrame()
testingdf = playerdf.loc[playerdf['gameid'] == "TRLH3/1000140076"]
```

```python
testingdf["impactscore"] = (testingdf["kills"]*3)-testingdf["deaths"]+(testingdf['assists']*1.5)+(testingdf["total cs"]*.02)
testingdf
```

Now that we know how to do the calculation on the testingdf, lets do it on the playerdf

```python
playerdf['impactscore'] = (playerdf["kills"]*3)-playerdf["deaths"]+(playerdf['assists']*1.5)+(playerdf["total cs"]*.02)
playerdf
```

Next thing we will do is talk about the quality of the impact the player has, this will using some new calculated values and some already available items. 

KP, KDA, D%, CSPM, DPM, DTPM, Damage Share, Gold Share, Total CS

After thinking about it, ill build both, and label them internal vs external, the internal score.

```python
testingdf["KDA"] = (testingdf["kills"]+testingdf["assists"]) / testingdf["deaths"]
testingdf["KP"] = (testingdf["kills"]+testingdf["assists"])  / testingdf["teamkills"]
testingdf["D%"] = testingdf["deaths"] / testingdf['teamdeaths']
testingdf
```

```python
playerdf["KDA"] = (playerdf["kills"]+playerdf["assists"]) / playerdf["deaths"]
playerdf["KP"] = (playerdf["kills"]+playerdf["assists"])  / playerdf["teamkills"]
playerdf["D%"] = playerdf["deaths"] / playerdf['teamdeaths']
playerdf
#playerdf = playerdf[playerdf["datacompleteness"]== "complete"].dropna(how = "any")
#playerdf.to_csv("ImpactScoreData.csv", index = False)
#This KDA score isn't correct for every line, I fix it in excel
```
