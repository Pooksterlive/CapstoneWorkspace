```python
matchdf = pd.read_csv("2022_LoL_esports_match.csv")
matchdf
```

```python
listid = matchdf['gameid'].tolist()
setid = set(listid)
playerdf = pd.DataFrame()
for item in setid:
    df_new = matchdf[matchdf["gameid"]== item]
    player = df_new.iloc[:10]
    playerdf = playerdf.append(player)
```

```python
playerdf
```

There are a number of columns that contain little to no data, so we are just going to remove them from the df. We would keep them if we were looking at team specific data.

```python
playerdf = playerdf.drop(["firstdragon", 'dragons','opp_dragons','elementaldrakes','opp_elementaldrakes','infernals','mountains','clouds','oceans','dragons (type unknown)','elders','opp_elders','firstherald','heralds','opp_heralds','firstbaron','barons','opp_barons','firsttower','towers','opp_towers','firstmidtower','firsttothreetowers','turretplates','opp_turretplates','inhibitors','opp_inhibitors'], axis = 1)
```

Now that the data is cleaned up, lets separate it by region, and only take the region we want to test on. In this case, 

```python
leagueset = list(playerdf["league"])
leagueset = set(leagueset)
mainregions = pd.DataFrame()
league = ['LCS']
for item in league:
    throwaway = playerdf[playerdf["league"] == item]
    mainregions = mainregions.append(throwaway)
    
```

```python
mainregions
```

Lets add the new metrics to this dataset

```python
mainregions["KDA"] = (mainregions["kills"]+mainregions["assists"]) / mainregions["deaths"]
mainregions["KP"] = (mainregions["kills"]+mainregions["assists"])  / mainregions["teamkills"]
mainregions["D%"] = mainregions["deaths"] / mainregions['teamdeaths']
mainregions["FantasyScore"] = (mainregions['kills']*3)-mainregions["deaths"]+(mainregions['assists']*1.5)+(mainregions["total cs"]*.02)
mainregions
```

Time to reduce the datset even more. There are alot of variables not usefull for the analysis, so removing the variables is best for efficiency

```python
variables = ['league','year','split','date','side',"playername",'position','teamname',"kills","deaths",'assists','doublekills','firstbloodkill'
             ,'firstbloodassist','firstbloodvictim','dpm','wpm','goldspent','total cs','minionkills','monsterkills',
             'monsterkillsenemyjungle','goldat15','xpat15','csat15','golddiffat15','xpdiffat15',
             'csdiffat15','killsat15','deathsat15','assistsat15','result','FantasyScore','KDA','KP','D%',"earnedgoldshare",
            'damageshare','cspm','damagetakenperminute','earnedgold','earned gpm']
#the commented section is if you are finding a null error in the factor reduction. In my case, the LCS 
#data doesn't have those issues, so I keep it commented out
reduced_df = mainregions[variables]#[mainregions["datacompleteness"]== 'complete'].dropna(how="any")
#reduced_df.to_csv("2022_Reduced.csv", index = False)
```

Now we can reduce the variables that we know have collinearity issues

```python
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#The first variable will be the @15 minute variables
features15 = ['goldat15','xpat15','csat15']
x = reduced_df.loc[:,features15].values
y = reduced_df.loc[:,['playername']].values
x = StandardScaler().fit_transform(x)
pca = PCA(n_components=1)
principalComponents = pca.fit_transform(x)
reduced_df['PCA15'] = principalComponents
pca.explained_variance_ratio_
```

```python
featureshare = ["earnedgoldshare",'damageshare']
x = reduced_df.loc[:,featureshare].values
y = reduced_df.loc[:,['playername']].values
x = StandardScaler().fit_transform(x)
pca = PCA(n_components=1)
principalComponents = pca.fit_transform(x)
reduced_df['PCAShare'] = principalComponents
pca.explained_variance_ratio_
```

```python
featureshare= ['earned gpm','cspm']
x = reduced_df.loc[:,featureshare].values
y = reduced_df.loc[:,["playername"]].values
x = StandardScaler().fit_transform(x)
#n_components should equal the total amount of components at the end
pca = PCA(n_components=1)
principalComponents = pca.fit_transform(x)
reduced_df["PCApm"] = principalComponents
pca.explained_variance_ratio_
```

```python
reduced_df['Counts'] = reduced_df.groupby(['playername'])['playername'].transform("count")
reduced_df
reduced_df.to_csv("2022_Reduced.csv", index = False)
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import ProbPlot
plt.style.use('seaborn') # pretty matplotlib plots
plt.rc('font', size=14)
plt.rc('figure', titlesize=18)
plt.rc('axes', labelsize=15)
```

```python
playerdf = pd.read_csv("PCATest1.csv")
playerdf = playerdf.rename(columns ={"total cs" : "total_cs"})#the statsmodel doesn't allow for spaces in variable names
```

```python
variables = ['kills','deaths','assists','PCA15','PCAShare','PCApm']
x = playerdf[variables].to_numpy() 
y = playerdf["result"]
```

```python
model = sm.Logit(y, sm.add_constant(x)).fit()
Y_pred = model.predict(sm.add_constant(x))
print_model = model.summary()
print(print_model)
```

Now that we have the original model, lets fit our 2022 data to this model

```python
variables = ['kills','deaths','assists','PCA15','PCAShare','PCApm'] # Sets X variables
x = reduced_df[variables].to_numpy() #makes the x variables an array, needed for the model
y = reduced_df["result"] # y variable
```

```python
red_model = sm.Logit(y, sm.add_constant(x)).fit() #Creates model
Y_pred = red_model.predict(sm.add_constant(x)) #produces predictions
print_model = red_model.summary() #creates the summary table
reduced_df['predscore'] = Y_pred # adds predictions to df
print(print_model)
```

Now that We have this new 2022 model, we can see the rank the teams based on the predicted scores

```python
predict_df = reduced_df.groupby('teamname', as_index = False)['predscore'].mean()
predict_df.sort_values(by=['predscore'])
```

We can do the same with players, and the player with the highest average score would be MVP

```python
#We need to drop players that played less than the half of the games in the regular season.
#This is how Riot Games 
mvp = reduced_df.groupby('Counts').filter(lambda x: len(x)>9)

mvp_df = mvp.groupby('playername',as_index = False)['predscore'].mean()
mvp_df.sort_values(by=['predscore'])
#mvp_df.to_csv("2022_MVP.csv",index = False)
```

According to this model, Bjergsen should have been the Spring Split MVP, if you include playoff games. Typically MVP is based only on regular season, but I like the regular + post season MVP discussion better.
