Now that we know how to get player data and team data seperated, we can run the loop to do this automatically. 

```python
#The first step is getting all of our data files loaded in
files = []
location = os.listdir(filepath)
for item in location:
    files.append(item)
```

```python
counter = 2014
for file in files:
    fulldata = pd.read_csv(file)
    listid = fulldata["gameid"].tolist()
    setid = set(listid)
    player_data = pd.DataFrame()
    team_data = pd.DataFrame()
    for item in setid:
        df_new = fulldata[fulldata['gameid'] == item]
        team = df_new.iloc[-2:]
        player = df_new.iloc[:10]
        player_data = player_data.append(player)
        team_data = team_data.append(team)
    print("")
    player_data.to_csv(str(counter)+"_player_data.csv",index = False)
    print("Player Data CSV Created")
    team_data.to_csv(str(counter)+"_team_data.csv",index = False)
    print("Team Data CSV Created")
    counter += 1  
```
